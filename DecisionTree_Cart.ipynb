```python
#Node Structure
#It'll be having 2 nodes - left, right
class Node:
    def __init__(self,attribute: list):
        self.attribute = attribute
        self.left: Node = None
        self.right: Node = None
```

```python
# Class implementation
class DecisionTree:
    def __init__(self):
        self.root: Node = None
        self.Gini_D = 0 

    def fit(self,df: pd.DataFrame):
        self.df = df
        # find the unique Target classes
        self.Target_classes = df['Actual'].unique()
        #Calculate the gini index...
        self.Gini_D = self.gini_dj(df)
        print('Gini_D: ',self.Gini_D)
        
        # Since root is none initially...
        self.root = Node(self.getAttribute(df))
        
    def constructTree(self,node: Node,df: pd.DataFrame):
        pass
    
    def gini_dj(self,df: pd.DataFrame):
        
        return  1 - np.sum([np.square(len(df[df['Actual']==cls])/len(df)) for cls in self.Target_classes])
        
    
    def getAttribute(self,df: pd.DataFrame):
        # No need to find gain if the datapoints has target classes with only yes/ only no as target class...
        cat = df['Actual'].unique()
        if len(cat)==1:
            return cat[0]

        # Find gini for each combination....
        gini_cat_dict = {}    # To find gini value for each combination
        gini_dict = {}        # To find gini value for each feature
        combination_dict = {} # To find the left and right edges...
        
        for feature in df.columns:
            if feature != 'Actual':
                cat_arr = df[feature].unique()
                # split the combinations....
                for categories in cat_arr:
                    sum_ = 0
                    # Left of combination ex: youth....
                    sum_ += len(df[df[feature]==categories])/len(df) * self.gini_dj(df[df[feature]==categories])
                    # Right of combinbation ex: M_A, Senior
                    sum_ += len(df[df[feature]!=categories])/len(df) * self.gini_dj(df[df[feature]!=categories])
                    # Store the gini value for each combinations
                    gini_cat_dict[categories] = self.Gini_D - sum_
                ##....
                print(gini_cat_dict)
                # Select the combination with highest gini value....
                combination_dict[feature] = max(gini_cat_dict, key = lambda k: gini_cat_dict[k])
                # Let the highest gini value be the gini value for feature ex: age..
                gini_dict[feature] = max(gini_cat_dict.values())
                gini_cat_dict = {}

        attrib = max(gini_dict, key = lambda k: gini_dict[k])
        # 1. Return the feature with highest gini
        # 2. Return the combination to keep track of left and right node..
        # Therefore return it as a dictionary....
        # Key - Attribute..
        # value - Combination
        return {attrib,combination_dict[attrib]}
            
                        
            
        
```
