```python
Y = df.target
X = df.drop(columns='target')
```

```python
class Madaline:
    def __init(self,threshold: float)->None:
        self.threshold_value = threshold
        
    def fit(self,X: pd.DataFrame,Y: pd.Series,l_r: float):
        self.X = X
        self.Y = Y
        self.getValues()
        # Hidden layer..
        self.weights_hidden = [] # weights of hidden layer w11,w21,......
        self.bias_hidden = [] # Bias of hidden layer b1, b2....
        # Output layer..
        self.weights_out = [] # Weights of output layer v1, v2, ......
        self.Z_in = [] 
        self.learning_rate = l_r
        count = 0
        for i in range(len(self.X)):
            self.hidden_input(i)
            print('Z matrix...')
            print(self.Z_in)
            self.Z_j = [self.threshold(val) for val in self.Z_in]
            print('Z matrix after threshold...')
            print(self.Z_j)
            self.output()
            print('Output_val :',self.output_val)
            while(self.Y.loc[i]!=self.output_val and count<=100):
                count+=1
                print('Count: ',count)
                self.update(self.Y.loc[i],i)
                self.hidden_input(i)
                print('Z matrix...')
                print(self.Z_in)
                self.Z_j = [self.threshold(val) for val in self.Z_in]
                print('Z matrix after threshold...')
                print(self.Z_j)
                self.output()
        
    def update(self,target,index: int):
        #index - Determines the data point number..
        #Updating weights for hidden layer...
        print('Updating hidden layer weights....')
        for i in range(len(X.columns)):
            for j in range(self.nodes_in_hidden_layer):
                # wij = wij + l_r* (target - Zinj) * xi
                self.weights_hidden[i][j] = self.weights_hidden[i][j] + self.learning_rate * (target - self.Z_in[j])*self.X.iloc[index][i]
                
        #Updating bias for hidden layer...
        for j in range(self.nodes_in_hidden_layer):
            # bj = bj + l_r* (target - Zinj) 
            self.bias_hidden[j] = self.bias_hidden[j] + self.learning_rate * (target - self.Z_in[j])
            
        print('Updated weights of hidden layer.....')
        print(self.weights_hidden)
        print('Updated bias of hidden layer....')
        print(self.bias_hidden)
        

    def output(self):
        # (1 x n) -> (n x 1)
        z = np.transpose(self.Z_j)
        
        # b + z1v1 + z2v2....zmvm
        # y_in -> (1 x 1)
        y_in = self.bias_output+np.dot(z,self.weights_out)
        print('y_in: ',y_in)
        
        self.output_val = self.threshold(y_in)


    def hidden_input(self,index: int):
        # (n x m) -> (m x n)
        weight_arr = np.transpose(self.weights_hidden)

        #(1 x n) -> (n x 1 )
        x = np.transpose(self.X.iloc[index])

        # (m x n) * (n x 1) -> m x 1
        # w11*x1 +w21 * x2 ......
        res = np.dot(weight_arr,x) 
        
        self.Z_in = [] # otherwise value will be appended in the previous zin array..
        #which results in varied dimension...
        for i in range(len(self.X.columns)):
            # b1+ w11 * x1 + w21 * x2
            self.Z_in.append(self.bias_hidden[i]+res[i])


    def threshold(self,val : float)->int:
        #Activation function..
        return 1 if val>=0 else 0
    
    def getValues(self):

        # no. of nodes in hidden layer....
        self.nodes_in_hidden_layer = int(input('No. of hidden layer elements: ')) 
        #Getting the weights for hidden layer..
        print('Enter the weights for hidden layer...')
        for i in range(len(self.X.columns)):
            temp = []
            for j in range(self.nodes_in_hidden_layer):
                temp.append(float(input())) # since weight is a 2d list...
            self.weights_hidden.append(temp)

        
        print('Enter the bias for the hidden layer.....')
        for i in range(len(self.X.columns)):
            self.bias_hidden.append(float(input()))

        print('Enter the weights for output layer...')
        # v1,v2,......
        for i in range(self.nodes_in_hidden_layer):
            self.weights_out.append(float(input()))
            

        self.bias_output = float(input('Enter the bias for output layer: '))
```

```python
model = Madaline()
model.fit(X,Y,0.6)
```
